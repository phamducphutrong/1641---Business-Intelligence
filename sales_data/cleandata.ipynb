```python
import csv
from string import ascii_uppercase as letters

# Tạo 1 list bảng chữ cái từ a-z
list_letters = []
for letter in letters:
    list_letters.append(letter)

# Tạo 1 list số từ 1 đến 26
list_numbers = list(range(1, 27))

fw = open('../sales_data/clean_data.csv', 'w', newline='')
writer = csv.writer(fw)
clean_data = []
with open('../sales_data/sales_data.csv', 'r') as fr:
    reader = csv.reader(fr)
    # bỏ qua dòng đầu tiên và gán list của dòng đầu tiên bằng header
    header = next(reader)
    # Đổi tên cột flag thành isBought
    header[0] = 'isBought'
    header[4] = 'MinAge'
    header.insert(5, 'MaxAge')
    writer.writerow(header)

    for row in reader:
        clean_data = row

        # Đổi Y thành Yes, N thành No tại cột isBought
        if clean_data[0] == 'Y':
            clean_data[0] = 'Yes'
        else:
            clean_data[0] = 'No'

        # Đổi M thành Male và F thành Female tại cột giới tính
        if clean_data[1] == 'M':
            clean_data[1] = 'Male'
        else:
            clean_data[1] = 'Female'

        # Đổi rỗng thành độc thân tại cột hôn nhân
        if clean_data[7] == '':
            clean_data[7] = 'Single'
        
        # Đổi U thành Unknown, Y thành Yes, N thành No tại cột Child
        if clean_data[8] == 'U':
            clean_data[8] = 'No'
        elif clean_data[8] == 'Y':
            clean_data[8] = 'Yes'
        else:
            clean_data[8] = 'No'

        # Đổi rỗng thành thuê nhà tại cột House-Owner
        if clean_data[11] == '':
            clean_data[11] = 'Renter'

        # Đổi rỗng thành Unknown, <HS thành near hish school, HS thành high school, Grad thành Graduated, Bach thành Bachelor tại cột Education
        if clean_data[2] == '':
            clean_data[2] = 'Unknown'
        else:
            clean_data[2] = clean_data[2].split('. ')[1]
        if clean_data[2] == '<HS':
            clean_data[2] = 'Near high school'
        elif clean_data[2] == 'HS':
            clean_data[2] = 'High school'
        elif clean_data[2] == 'Grad':
            clean_data[2] = 'Graduated'
        elif clean_data[2] == 'Bach':
            clean_data[2] = 'Bachelor'

        # Đổi cột age thành min age và thêm 1 cột max age
        # Đầu tiên split cột age và lấy index 0. Sau đó gán lại giá trị của cột age bằng với giá trị của index 0 vừa nhận được
        clean_data[4] = clean_data[4].split('_')[0]
        # Kiểm tra xem giá trị của cột age hiện tại xem bằng bao nhiêu.
        # Giá trị của cột age bây giờ sẽ thuộc vào khoảng từ 1 đến 7.
        # Với mỗi giá trị sẽ có một min age và max age riêng.
        # gán cột age hiện tại bằng có trị min age và chèn thêm 1 cột max age.
        # chú ý: hàm insert có 2 tham số trong đó tham số đầu tiên là vị trí index, tham số thứ hai là giá trị được gán.
        if clean_data[4] == '1':
            clean_data[4] = 1
            clean_data.insert(5, 100)
        elif clean_data[4] == '2':
            clean_data[4] = 1
            clean_data.insert(5, 25)
        elif clean_data[4] == '3':
            clean_data[4] = 26
            clean_data.insert(5, 35)
        elif clean_data[4] == '4':
            clean_data[4] = 36
            clean_data.insert(5, 45)
        elif clean_data[4] == '5':
            clean_data[4] = 46
            clean_data.insert(5, 55)
        elif clean_data[4] == '6':
            clean_data[4] = 56
            clean_data.insert(5, 65)
        elif clean_data[4] == '7':
            clean_data[4] = 66
            clean_data.insert(5, 100)

        # Đổi cột fam_income thành số
        clean_data[-1] = list_numbers[list_letters.index(clean_data[-1])]

        # Xóa số, chỉ giữ lại chữ, nếu như chữ là med thì đổi thành medium tại cột Mortgage
        clean_data[-5] = clean_data[-5][1:]
        if clean_data[-5] == 'Med':
            clean_data[-5] = 'Medium'
        
        # Đổi Y thành Yes, N thành No tại cột Online
        if clean_data[6] == 'Y':
            clean_data[6] = 'Yes'
        else:
            clean_data[6] = 'No'

        # Ghi list 'clean_data' vào file clean_data.csv
        writer.writerow(clean_data)
fw.close()
```
