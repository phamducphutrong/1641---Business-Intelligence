```python
import matplotlib.pyplot as plt
import numpy  as np

# create 100 linearly spaced numbers from -10 to 10
x = np.linspace(-10, 10, 100)
y = 3 * x ** 2 - 2 * x + 1 # y = 3x^2 - 2x + 1

# draw the plot
plt.plot(x, y, 'r')

# set the x axis label
plt.xlabel('x')
# set the y axis label
plt.ylabel('y')

# set the plot title
plt.title('y = 3x^2 - 2x + 1')

# show the plot
plt.show()
```
